```python
# how do we account for user error eg: types "rainn" instead of "rain"
import difflib
from difflib import SequenceMatcher
SequenceMatcher(None,"rainn","rain").ratio()
```

```python
#we're going to use another method = get_close
from difflib import get_close_matches
help(get_close_matches)
```

```python
w_list = data.keys()
get_close_matches("rainn",w_list)
```

```python
def define(w):
    w = w.lower()
    if w in data:
        return(data[w]) #if the word is in there, then print out the meaning
    elif len(get_close_matches("rainn",w_list,cutoff=0.8)) > 0:#check if the word is in the similar words list
        temp = get_close_matches("rainn",w_list,cutoff=0.8)[0]
        yn = lower(input("Did you mean %s? Type 'y' if Yes, and 'n' if No"%(temp)))
        # prompt the user for the most likely match       
        if yn == "y":
            return data[temp]
        elif yn == "n":
            return "Word not found. Please try again."
        else:
            return "I'm not sure what you mean. Please try again."
    else:
        return "Word not found. Please try again."


define("Rain")
```
