# Stack Plot or Area Chart

Note: if you don't force the colour, it will just cycle through automatically

```python
year  = list(range(0,10))

taxes = [17,18,40,43,44,8,43,32,39,30]
overhead = [30,22,9,29,17,12,14,24,49,35]
entertainment = [41,32,27,13,19,12,22,18,28,20]

# making a fake legend
plt.plot([],[],label = "taxes",color ="m")
plt.plot([],[],label = "overhead",color = "c")
plt.plot([],[],label  = "entertainment",color = "b")

plt.legend()

plt.stackplot(year,taxes,overhead,entertainment, colors = ["m","c","b"])
```

# Pie Charts/Sector Graphs

```python
#Note that labels is a tuple

labels  = "Taxes","Overhead","Entertainment"

sizes = [25,32,12]

colours = ["c","m","b"]

plt.pie(sizes,labels=labels,colors=colours,startangle=90,shadow = True,autopct = '%1.1f%%',explode = (0.05,0.05,0.05))
plt.axis("equal")

    
```

# Reading from Files

```python
import csv

x = []
y = []

with open("F:/Google Drive/JupyterNotebooks/read_csv_example.txt","r") as csvfile:
    plots = csv.reader(csvfile,delimiter =",")
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
plt.plot(x,y,label = "Loaded from File")
plt.xlabel("Plot Number")
plt.ylabel("Other Plot Number")
plt.legend()
plt.title("Yay! I did it!")

    
```
