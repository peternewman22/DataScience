```python
with open(path, "r", encoding='utf8') as fh:
    data = json.load(fh)
print(data)
```

```python
class Subscription:
    def __init__(self, title, videoCount, published, description):
        self.title = title
        self.videoCount = videoCount
        self.published = published
        self.description = description
        
    def row(self):
        return {'title' : self.title, 'videoCount' : self.videoCount, 'published' : self.published, 'description' : self.description}

clean_data = []
for x in data:
    new = Subscription(x['snippet']['title'], 
                       x['contentDetails']['totalItemCount'], 
                       x['snippet']['publishedAt'], 
                       x['snippet']['description'])
    clean_data.append(new.row())
```

```python
df = pd.DataFrame(clean_data)
```
